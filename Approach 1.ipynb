```python
import pandas as pd

# Load the NIFTY50_all.csv file again
nifty50_df = pd.read_csv(os.path.join(extraction_path, 'NIFTY50_all.csv'))

# Display the first few rows of the dataframe
nifty50_df.head()
```

```python
# Reshape the data to have dates as rows and symbols as columns with their respective closing prices
pivot_df = nifty50_df.pivot(index='Date', columns='Symbol', values='Close')

# Display the first few rows of the reshaped dataframe
pivot_df.head()
```

```python
# Drop columns with more than 5% missing values
threshold = 0.05 * len(pivot_df)
filtered_df = pivot_df.dropna(axis=1, thresh=threshold)

# Forward fill to handle any remaining NaNs in the middle of the time series
filtered_df = filtered_df.fillna(method='ffill')

# If there are still NaNs left, it means they are at the beginning of the time series. We'll backward fill those.
filtered_df = filtered_df.fillna(method='bfill')

# Display the first few rows of the cleaned dataframe
filtered_df.head()
```

```python
import numpy as np

# Function to compute asset returns 
def StockReturnsComputing(StockPrice, Rows, Columns):
    StockReturn = np.zeros([Rows-1, Columns])
    for j in range(Columns):        # j: Assets
        for i in range(Rows-1):     # i: Daily Prices
            StockReturn[i,j] = ((StockPrice[i+1, j] - StockPrice[i,j]) / StockPrice[i,j]) * 100
    return StockReturn

# Convert the dataframe to numpy array
stockPriceArray = filtered_df.values
[Rows, Cols] = stockPriceArray.shape

# Compute stock returns
stockReturns = StockReturnsComputing(stockPriceArray, Rows, Cols)
stockReturns_df = pd.DataFrame(stockReturns, columns=filtered_df.columns, index=filtered_df.index[1:])

# Display the daily returns for the first few dates
stockReturns_df.head()
```

```python
# Compute mean returns and variance-covariance matrix of returns
meanReturns = stockReturns_df.mean()
covReturns = stockReturns_df.cov()

meanReturns, covReturns
```

```python
# Compute the market returns, which is the average return of all stocks
marketReturns = stockReturns_df.mean(axis=1)

# Compute betas for each stock
beta = {}
Var = marketReturns.var()
for symbol in stockReturns_df.columns:
    CovarMat = np.cov(stockReturns_df[symbol], marketReturns)
    Covar  = CovarMat[0,1]
    beta[symbol] = Covar / Var

beta_series = pd.Series(beta)
beta_series
```

```python
# Define equal weights for all stocks
num_stocks = len(filtered_df.columns)
weights = np.full((num_stocks,), 1/num_stocks)

# Compute mean and covariance of asset returns
meanReturns = stockReturns_df.mean()
covReturns = stockReturns_df.cov()

# Compute portfolio risk
portfolioRisk = np.matmul((np.matmul(weights, covReturns)), np.transpose(weights))

# Annualize portfolio risk for trading days = 251
annualizedRisk = np.sqrt(portfolioRisk * 251) 

# Compute expected portfolio return
portfolioReturn = np.matmul(np.array(meanReturns), weights.T)

# Annualize expected portfolio return
annualizedReturn = 251 * np.array(portfolioReturn) 

# Compute portfolio beta
portfolioBeta = np.matmul(beta_series.values, weights.T)

annualizedRisk, annualizedReturn, portfolioBeta
```

```python
import matplotlib.pyplot as plt

# Number of portfolios to simulate
num_portfolios = 10000

# Empty lists to store the portfolio performance metrics
portfolio_returns = []
portfolio_volatilities = []
portfolio_weights = []

# Loop through and simulate a large number of random portfolio combinations
for i in range(num_portfolios):
    weights = np.random.random(num_stocks)
    weights /= np.sum(weights)
    portfolio_weights.append(weights)
    portfolio_return = np.sum(meanReturns * weights)
    portfolio_returns.append(portfolio_return)
    portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(covReturns, weights)))
    portfolio_volatilities.append(portfolio_stddev)

# Convert lists to arrays
portfolio_returns = np.array(portfolio_returns)
portfolio_volatilities = np.array(portfolio_volatilities)

# Plot the efficient frontier
plt.figure(figsize=(12, 8))
plt.scatter(portfolio_volatilities, portfolio_returns, c=portfolio_returns/portfolio_volatilities, marker='o', cmap='YlGnBu')
plt.colorbar(label='Sharpe Ratio')
plt.title('Efficient Frontier with Simulated Portfolios')
plt.xlabel('Portfolio Risk (Standard Deviation)')
plt.ylabel('Expected Portfolio Return')
plt.grid(True)
plt.show()
```

```python
import plotly.figure_factory as ff

# Pivot the data to get closing prices for all stocks
pivot_df = nifty50_df.pivot(index='Date', columns='Symbol', values='Close')

# Calculate the daily returns
daily_returns = pivot_df.pct_change()

# Calculate the correlation matrix
correlation_matrix = daily_returns.corr()

# Drop rows and columns with all NaN values
correlation_matrix = correlation_matrix.dropna(axis=0, how='all').dropna(axis=1, how='all')

# Fill any remaining NaN values with 0
correlation_matrix = correlation_matrix.fillna(0)

# Convert the index and columns to lists
x_values = correlation_matrix.columns.tolist()
y_values = correlation_matrix.index.tolist()

# Create a heatmap with increased figure size and adjusted font size
fig = ff.create_annotated_heatmap(z=correlation_matrix.values, x=x_values, y=y_values, 
                                  colorscale='Viridis', showscale=True, font_colors=["white", "black"],
                                  hoverinfo='z')

fig.update_layout(title='Correlation Heatmap of Daily Returns of NIFTY 50 Stocks',
                  width=1000, height=1000,  # Increase figure size
                  autosize=False)

# Adjust font size for x and y axis tick labels
fig.update_xaxes(tickfont=dict(size=10))
fig.update_yaxes(tickfont=dict(size=10))

fig.show()
```
